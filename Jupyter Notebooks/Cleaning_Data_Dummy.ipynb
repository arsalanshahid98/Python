```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df = pd.DataFrame(people)

# replacing custom missing values with default nunpy missing NaN values
df.replace("NA", np.nan, inplace = True)
df.replace("Missing", np.nan, inplace = True)
df.replace("None", np.nan, inplace = True) 
```

```python
df
```

```python
df.dropna() # deletest all the values with NaN in it
```

```python
df.dropna(axis = "index", how = "any") # same thing as above but with default arguments
```

```python
df.dropna(axis = "index", how = "all") # drops rows if all the values in a row are missing which is all NaN and none
```

```python
df.dropna(axis = "columns", how = "all") # drops columns if all are NaN in columns
```

```python
df.dropna(axis = "columns", how = "any") #drops any columns with NaN values
```

```python
df.dropna(axis = "index", how = "any", subset = ["email"]) # drops rows if there is NaN in specific columns ie email
```

```python
df.isna() # checking if any value is na or not
```

```python
df.fillna("MISSING") # replacing all na values with something
```

```python
df.fillna(0)
```
